```python
import math

def probability_exact(num_people, pho, system_capacity):
    result = (pho**num_people) * ((1 - pho) / (1 - (pho**(system_capacity + 1))))
    return result

def probability_at_least(num_people, pho, system_capacity):
    summation_res = 0
    for i in range(0, num_people):
        summation_res += (pho**i) * ((1 - pho) / (1 - (pho**(system_capacity + 1))))
    result = 1 - summation_res
    return result

def probability_at_most(num_people, pho, system_capacity):
    summation_res = 0
    for i in range(0, num_people + 1):
        summation_res += (pho**i) * ((1 - pho) / (1 - (pho**(system_capacity + 1))))
    result = summation_res
    return result

def main():
    # Input parameters
    system_capacity = int(input("What is the maximum limit of people that can be in the system? "))
    arrival_rate = float(input("What is the arrival rate (people per time unit) in your system? "))
    service_rate = float(input("What is the service rate (people per time unit) in your system? "))
    pho = arrival_rate / service_rate

    # Calculate metrics
    Ls = (pho * ((1 + (system_capacity * (pho**(system_capacity + 1)))) - ((system_capacity + 1) * (pho**system_capacity)))) / ((1 - pho) * (1 - (pho**(system_capacity + 1))))
    arr_eff = arrival_rate * (1 - probability_exact(system_capacity, pho, system_capacity))
    Lq = Ls - (arr_eff / service_rate)
    Ws = Ls / arr_eff
    Wq = Lq / arr_eff

    # Display system metrics
    print(f"The expected length of the system is {round(Ls, 4)} people")
    print(f"The expected length of the queue is {round(Lq, 4)} people")
    print(f"The expected waiting time of the system is {Ws}")
    print(f"The expected waiting time of the queue is {Wq} ")

    # Query probabilities
    c = "yes"
    while c.lower() == "yes":
        num_people = int(input("\nHow many people in the system are you querying the probability for? "))
        option = int(input(f"Which probability are you looking for?\n1. Exactly {num_people} people\n2. At least {num_people} people\n3. At most {num_people} people\n"))

        if option == 1:
            print(f"The probability of exactly {num_people} people in the system is {probability_exact(num_people, pho, system_capacity) * 100:.2f}%")
        elif option == 2:
            print(f"The probability of at least {num_people} people in the system is {probability_at_least(num_people, pho, system_capacity) * 100:.2f}%")
        elif option == 3:
            print(f"The probability of at most {num_people} people in the system is {probability_at_most(num_people, pho, system_capacity) * 100:.2f}%")
        else:
            print("Invalid option!")

        c = input("Do you want to query again? (yes/no) ")

    print("End")

if __name__ == "__main__":
    main()
```
